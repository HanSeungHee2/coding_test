길이가 같은 두 개의 큐가 주어집니다. 하나의 큐를 골라 원소를 추출(pop)하고, 추출된 원소를 다른 큐에 집어넣는(insert) 작업을 통해 각 큐의 원소 합이 같도록 만들려고 합니다. 이때 필요한 작업의 최소 횟수를 구하고자 합니다. 한 번의 pop과 한 번의 insert를 합쳐서 작업을 1회 수행한 것으로 간주합니다.

큐는 먼저 집어넣은 원소가 먼저 나오는 구조입니다. 이 문제에서는 큐를 배열로 표현하며, 원소가 배열 앞쪽에 있을수록 먼저 집어넣은 원소임을 의미합니다. 즉, pop을 하면 배열의 첫 번째 원소가 추출되며, insert를 하면 배열의 끝에 원소가 추가됩니다. 예를 들어 큐 [1, 2, 3, 4]가 주어졌을 때, pop을 하면 맨 앞에 있는 원소 1이 추출되어 [2, 3, 4]가 되며, 이어서 5를 insert하면 [2, 3, 4, 5]가 됩니다.

다음은 두 큐를 나타내는 예시입니다.

queue1 = [3, 2, 7, 2]
queue2 = [4, 6, 5, 1]
두 큐에 담긴 모든 원소의 합은 30입니다. 따라서, 각 큐의 합을 15로 만들어야 합니다. 예를 들어, 다음과 같이 2가지 방법이 있습니다.

queue2의 4, 6, 5를 순서대로 추출하여 queue1에 추가한 뒤, queue1의 3, 2, 7, 2를 순서대로 추출하여 queue2에 추가합니다. 그 결과 queue1은 [4, 6, 5], queue2는 [1, 3, 2, 7, 2]가 되며, 각 큐의 원소 합은 15로 같습니다. 이 방법은 작업을 7번 수행합니다.
queue1에서 3을 추출하여 queue2에 추가합니다. 그리고 queue2에서 4를 추출하여 queue1에 추가합니다. 그 결과 queue1은 [2, 7, 2, 4], queue2는 [6, 5, 1, 3]가 되며, 각 큐의 원소 합은 15로 같습니다. 이 방법은 작업을 2번만 수행하며, 이보다 적은 횟수로 목표를 달성할 수 없습니다.
따라서 각 큐의 원소 합을 같게 만들기 위해 필요한 작업의 최소 횟수는 2입니다.

길이가 같은 두 개의 큐를 나타내는 정수 배열 queue1, queue2가 매개변수로 주어집니다. 각 큐의 원소 합을 같게 만들기 위해 필요한 작업의 최소 횟수를 return 하도록 solution 함수를 완성해주세요. 단, 어떤 방법으로도 각 큐의 원소 합을 같게 만들 수 없는 경우, -1을 return 해주세요.

리스트 클래스의 pop메서드는 리스트에서 요소를 꺼내면서 제거하는 메서드이다. 디큐과정은 무조건 먼저들어간 요소가 나와야한다. list.pop(0): 0번째 인덱스값(먼저 들어간값)을 꺼내고 제거.

리스트 클래스의 insert는 특정 위치에 요소를 끼워넣는 메서드.

```python
queue1 = [1]*10 + [10]
queue2 = [2]*10

# queue3 = queue1은 주소복사, 즉 두 변수는 같은 메모리공간을 가리킨다. queue3 - queue1.copy()를 써야한다.
#list.pop(index): index에 해당하는 원소 삭제.
	# [1, 2, 1, 2], [1, 10, 1, 2] 을 풀어보자.

#재귀함수로 풀수있을거같은데.. 큰거에서 작은거로 반복하는거지.
def test1(queue1, queue2):
    sum1 = sum(queue1)
    sum2 = sum(queue2)

    all_components = queue1 + queue2
    work_count = 0

    if sum2 == sum1:
        yield queue1, queue2, work_count
        return

    elif max(all_components) > (sum(all_components) - max(all_components)):
        yield -1
        return

    else:
        while sum2 != sum1:
            # 👉 여기서 상태 하나씩 뽑기
            yield queue1.copy(), queue2.copy(), work_count

            if sum2 > sum1:
                return_value = queue2.pop(0)
                queue1.append(return_value)
                work_count += 1

            else:
                return_value = queue1.pop(0)
                queue2.append(return_value)
                work_count += 1

            sum1 = sum(queue1)
            sum2 = sum(queue2)

    yield queue1, queue2, work_count

# 이터레이터 선언!
output = test1(queue1, queue2)

for i in range(10):
    print(next(output))
```

```python
def find_cases(arr):
    n = len(arr)
    result = []

    def dfs(idx, A, B):
        if idx == n:
            if A and B and sum(A) == sum(B):
                result.append((A[:], B[:]))
            return

        # 1. A에 넣기
        A.append(arr[idx])
        dfs(idx+1, A, B)
        A.pop()

        # 2. B에 넣기
        B.append(arr[idx])
        dfs(idx+1, A, B)
        B.pop()

        # 3. 안 넣기
        dfs(idx+1, A, B)

    dfs(0, [], [])
    return result if result else None
```
